```python
import numpy as np
import pandas as pd
```

```python
streaming_df = pd.read_csv('streaming_set.csv')
streaming_df.head(3)
```

```python
'''Elegimos un DataFrame de prueba más pequeño debido a la capacidad de mi compu '''
df =streaming_df[streaming_df['platform']=='netflix'] 
df.head(3)
```

```python
df.reset_index(inplace=True)
df.head(5)
```

```python
dicc_indices_movieid = df['id'].to_dict()
dicc_indices_movieid
```

```python
dicc_movieid_indice = {valor: clave for clave, valor in dicc_indices_movieid.items()}
dicc_movieid_indice
```

```python
import re

def quitar_numeros(argumento):
    s = argumento.lower()
    s = re.sub(r"\d+", "", s)
    return s
```

```python
''' En cada texto, cuantas veces está cada palabra '''
from sklearn.feature_extraction.text import CountVectorizer

contador_argumento = CountVectorizer(preprocessor=quitar_numeros, min_df=5)

argumentos_bag_of_words= (contador_argumento.fit_transform(df["description"]).toarray())

argumentos_bag_of_words
```

```python
argumentos_bag_of_words.shape
```

```python
''' Prsentamos como una dataframe el array anterior '''
columnas_argumentos = [tup[0] for tup in sorted(contador_argumento.vocabulary_.items(), key = lambda x:x[1])]

argumentos_bag_of_words_df = pd.DataFrame(argumentos_bag_of_words,columns=columnas_argumentos,index=df['title'])
argumentos_bag_of_words_df.head()
```

```python
''' Función que forma pares de combinaciones de una '''
from itertools import combinations

def tokerizador_generos(string_generos):
    generos_separados = string_generos.split(",")
    resultado = []
    for tamaño in [1,2]:
        combs = ["Generos - " + ",".join(sorted(tupla))
                 for tupla in combinations(generos_separados, r = tamaño)]
        resultado = resultado + combs
    return sorted(resultado)
```

```python
''' Convertimos los géneros a más columnas '''

from sklearn.feature_extraction.text import CountVectorizer

contador_generos = CountVectorizer(tokenizer=tokerizador_generos,
                                   token_pattern=None,
                                   lowercase=False)
contador_generos.fit(df['listed_in'])

generos_bag_of_words =  contador_generos.fit_transform(df['listed_in']).toarray()

generos_bag_of_words
```

```python
columnas_generos = [tup[0] for tup in
                    sorted(contador_generos.vocabulary_.items(),
                           key=lambda x: x[1])]
generos_bag_of_words_df = pd.DataFrame(generos_bag_of_words,
                                       columns=columnas_generos,
                                       index=df['title'])

generos_bag_of_words_df 
```

```python
bag_of_words_ambos =np.hstack((argumentos_bag_of_words, generos_bag_of_words))

bag_of_words_ambos_df = pd.DataFrame(bag_of_words_ambos,
                                     columns=columnas_argumentos+columnas_generos,
                                     index=df['title'])
bag_of_words_ambos_df
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tf_idf = TfidfTransformer()

tf_idf_pelis = tf_idf.fit_transform(bag_of_words_ambos_df).toarray()

tf_idf_pelis
```

```python
tf_idf_pelis_df = pd.DataFrame(tf_idf_pelis,
                               index=df['title'],
                               columns=columnas_argumentos+columnas_generos)
tf_idf_pelis_df
```

```python
''' Matriz de similaridades '''
from sklearn.metrics.pairwise import cosine_similarity

cosine_sims = cosine_similarity(tf_idf_pelis_df)

matriz_similaridades_df = pd.DataFrame(cosine_sims,
                                       index=df['title'],
                                       columns=df['title'])
matriz_similaridades_df
```

```python
''' Eliminamos la diagonal '''
np.fill_diagonal(matriz_similaridades_df.values, np.nan)

matriz_similaridades_df.head()
```

```python
orden_pelis_cosine_sims_por_fila = np.argsort((-cosine_sims), axis=1)
orden_pelis_cosine_sims_por_fila
```

```python
cosine_sims_oredanas = np.sort(-cosine_sims, axis = 1)

cosine_sims_oredanas
```

```python
def top_k_similares(id, k):
    fila_cosine_sims = dicc_movieid_indice[id]

    lista_ordenada_pelis_sims = orden_pelis_cosine_sims_por_fila[fila_cosine_sims]
    lista_ordenada_sims = cosine_sims_oredanas[fila_cosine_sims]

    top_k = lista_ordenada_pelis_sims[:k]
    cosine_sims_top_k = lista_ordenada_sims[:k]

    top_k_df = df.loc[top_k].copy()
    top_k_df['similaridad'] = cosine_sims_top_k

    return top_k_df
```

```python
ratings_df = pd.read_csv('ratings_set.csv')
ratings_df.head(5)
```

```python
ratings_netflix_df=ratings_df[ratings_df['movieId'].str.contains('ns')]
ratings_netflix_df.head()
```
